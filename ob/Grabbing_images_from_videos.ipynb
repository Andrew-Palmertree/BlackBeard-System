```python
import cv2
import os

def extract_frames(video_path, output_folder, interval=3):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Initialize variables
    frame_count = 0
    success = True
    
    while success:
        # Set frame position to extract every 'interval' seconds
        cap.set(cv2.CAP_PROP_POS_MSEC, (frame_count * 100 * interval))
        
        # Read the frame
        success, frame = cap.read()
        
        if success:
            # Save the frame to file
            output_path = os.path.join(output_folder, f"2frame_{frame_count}.jpg")
            cv2.imwrite(output_path, frame)
            print(f"Frame {frame_count} saved as {output_path}")
            frame_count += 1
    
    # Release the video capture object
    cap.release()
    cv2.destroyAllWindows()

# Example usage
video_path = r"C:\Users\andre\OneDrive\Pictures\OLD\Camera Roll\WIN_20240419_10_28_09_Pro.mp4"
output_folder = r"C:\Users\andre\OneDrive\Pictures\OLD\Camera Roll\Photos\Video_Output\Images"
interval = 3
extract_frames(video_path, output_folder, interval)
```

```python
import numpy as np
import cv2
import os
import random

# Load the image
image_path = r"C:\Users\andre\Downloads\11.png"
image = cv2.imread(image_path)

# Define augmentation functions
def gaussian_blur(image):
    return cv2.GaussianBlur(image, (5, 5), 0)

def rotate(image):
    angle = random.randint(-45, 45)
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    return cv2.warpAffine(image, rotation_matrix, (width, height))

def brightness_adjustment(image):
    alpha = random.uniform(0.5, 1.5)
    beta = random.randint(-50, 50)
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted

def scale(image):
    scale_factor = random.uniform(0.8, 1.2)
    height, width = image.shape[:2]
    new_height, new_width = int(height * scale_factor), int(width * scale_factor)
    return cv2.resize(image, (new_width, new_height))

def translate(image):
    x_shift = random.randint(-50, 50)
    y_shift = random.randint(-50, 50)
    height, width = image.shape[:2]
    translation_matrix = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
    return cv2.warpAffine(image, translation_matrix, (width, height))

def contrast_adjustment(image):
    alpha = random.uniform(0.5, 1.5)
    beta = random.randint(-50, 50)
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted

# Define augmentation sequence
augmentation_functions = [
    gaussian_blur,
    rotate,
    brightness_adjustment,
    scale,
    translate,
    contrast_adjustment
]

# Create directory to save augmented images
output_dir = r"C:\Users\andre\OneDrive\Pictures\OLD\Camera Roll\Photos\Video_Output\Images"
os.makedirs(output_dir, exist_ok=True)

# Augment and save 1000 images
for i in range(100):
    augmented_image = image.copy()
    for aug_func in augmentation_functions:
        augmented_image = aug_func(augmented_image)
    output_path = os.path.join(output_dir, f"augmented_11_image_{i}.jpg")
    cv2.imwrite(output_path, augmented_image)

print("Augmentation completed. 100 images created.")
```
